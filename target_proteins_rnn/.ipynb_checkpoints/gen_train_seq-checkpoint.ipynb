```python
import numpy as np
import pickle
import itertools as itool
```

```python
with open('../target_features/target_proteins/train_target_seq.pkl', 'rb') as f:
    train_target_seq = pickle.load(f)
    
with open('../target_features/train_target_IDs.pkl', 'rb') as f:
    train_target_IDs = pickle.load(f)
    
with open('../target_features/target_proteins/train_target_indices.pkl', 'rb') as f:
    train_target_indices = pickle.load(f)
```

```python
train_data_seq = []
for dbid, seq in train_target_seq.items():
    train_data_seq.append(seq)

with open('output/train_data_seq.pkl', 'wb') as f:
    pickle.dump(train_data_seq, f)
    
    
#print(train_data_seq[0])
all_seq = list(itool.chain.from_iterable(train_data_seq))

with open('output/train_all_seq.pkl', 'wb') as f:
    pickle.dump(all_seq, f)

print(all_seq[0:10])
print(len(all_seq))
lengths = np.zeros(len(all_seq))
for idx in range(len(all_seq)):
    lengths[idx] = len(all_seq[idx])

print("Largest length of all sequences:{}".format(np.max(lengths)))
max_idx = np.argmax(lengths)
print("Longest sequence:{}".format(all_seq[max_idx]))

```

```python
train_data_targets = []
idx = 0
for dbid, target in train_target_IDs.items():
    train_data_targets.append(target)
    if idx == 35:
        print(dbid)
    idx += 1
    
print("train_data_targets:{}, length:{}".format(train_data_targets, len(train_data_targets)))

with open('output/train_data_targets.pkl', 'wb') as f:
    pickle.dump(train_data_targets, f)
```

```python
train_data_target_indices = []
for dbid, index in train_target_indices.items():
    train_data_target_indices.append(index)
    
print("train_data_target_indices:{}, length:{}".format(train_data_target_indices, len(train_data_target_indices)))

with open('output/train_data_target_indices.pkl', 'wb') as f:
    pickle.dump(train_data_target_indices, f)
```

```python
with open('output/train_data_targets.pkl', 'rb') as f:
    train_data_targets = pickle.load(f)

with open('output/train_data_seq.pkl', 'rb') as f:
    train_data_seq = pickle.load(f)
    
print(train_data_targets[35])
print(train_data_seq[35])
```
